```python
import numpy as np 
from matplotlib import pyplot as plt 
from scipy import fft as fft

import matplotlib
plt.rcParams['font.sans-serif'] = ['KaiTi'] 
matplotlib.rcParams['axes.unicode_minus'] =False

%matplotlib inline
```

```python
def plot_f(t,f, title='时域'):
    plt.plot(t,f)
    plt.title(title)
    plt.xlabel("时间t")
    plt.ylabel("幅度f")
    
def plot_F(n, F, title='频域'):
    plt.plot(n,F)
    plt.title(title)
    plt.xlabel("频率f")
    plt.ylabel("幅度F")
```

```python

t = np.linspace(0, 1, 500, endpoint=False)
f = 0.5 * np.sin(2 * np.pi * 5 * t) + np.sin(2 * np.pi * 10 * t)
```

```python
plot_f(t, f, title='时域')
```

```python
def get_fft(t, f): 
    n = fft.fftfreq(len(t), 1/len(t))
    F = fft.fft(f)
    return n, F

n, F = get_fft(t, f)
plot_F(n,abs(F))
```

```python
plt.subplot(2,1,1)
plot_f(t, f, title='时域')
plt.subplot(2,1,2)
plot_F(n, abs(F), title='频域')
```

```python
ifft_f = fft.ifft(F)
plot_f(t, ifft_f, title='逆变换')
```
